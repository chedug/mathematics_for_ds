## Задача 1

We got:

\begin{align}
M(x) &= \sum x_i p_i \\
&= -2*0.2-1*0.1+0*0.2+p_4+2p_5 \\
&= -0.5 + p_4 +2p_6 \\
&= 0.1 \\
\\
\Rightarrow & p_4 +2p_5 = 0.6 \\
\\
\end{align}


And we also use the fact that $\sum p_i = 1$:

$$
0.2+0.1+0.2+p_4+p_5 = 1
$$

$$
p_4 + p_5 = 0.5
$$

So we got linear system of equations of the form Ax = b with

$$
A = \begin{bmatrix}
1 & 2 \\
1 & 1
\end{bmatrix}, \quad
b = \begin{bmatrix}
0.6 \\
0.5
\end{bmatrix}
$$

where x is the vector with probabilities $p_4$ and $p_5$:
$$
x = \begin{bmatrix}
p_4 \\
p_5
\end{bmatrix}
$$

```python
import numpy as np

A = np.array([[1, 2],
              [1, 1]])
b = np.array([0.6, 0.5])

x = np.linalg.solve(A, b)
print("Solution x:")
print(x)
print(f"p4 is {x[0]:.3f}")
print(f"p5 is {x[1]:.3f}")
```

Now, we need to find D(x). We know that $\mathbb{D}[x]= \mathbb{E}[X^2] - \mathbb{E}[X]^2$:

```python
x = np.array([-2,-1,0,1,2])
p = np.array([0.2,0.1,0.2,0.4,0.1])
x_squared = x ** 2
# E[X^2]

expectation_squares = np.dot(x_squared, p)


variance = expectation_squares - 0.1 ** 2

print(f"variance is: {variance:.4f}")
```

c) find and plot distriubtion function F(x)

```python
import numpy as np
import matplotlib.pyplot as plt

x_values = np.array([-2, -1, 0, 1, 2])
probabilities = np.array([0.2, 0.1, 0.2, 0.4, 0.1])

cumulative_probabilities = np.cumsum(probabilities)

plt.step(x_values, cumulative_probabilities, where='post', label='CDF')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('Cumulative Distribution Function (CDF)')
plt.grid()
plt.legend()
plt.show()
```

## Задача 6

image.png



Пусть $X_i$ - случайная величина равная тому, что станок $i$ работает спустя час. $q_i := 1-p_i$ вероятность того, что станок $i$ работает спустя час. Очевидно, $X_i \sim Ber(q_i)$.

Нас интересует случаная величина $Z:=X_1+X_2+X_3$, равная количеству работающих станков спустя час. $Z \in \{0,1,2,3\}$. Так как случайная величина дискретная, можем самостоятельно посчитать вероятности:


\begin{align*}
\mathbb{P}(Z=0) &= p_1 p_2 p_3 = 0.003\\
\\
\mathbb{P}(Z=1) &= \sum_{\substack{i+j+k=1}} \mathbb{P}(X_1=i, X_2=j, X_3=k) \\
&= q_1p_2p_3 + p_1q_2p_3+p_1p_2q_3 \\
&= 0.056\\
\\
\mathbb{P}(Z=2) &= \sum_{\substack{i+j+k=2}} \mathbb{P}(X_1=i, X_2=j, X_3=k) \\
&= p_1q_2q_3 + q_1p_2q_3 + q_1q_2p_3 \\
&= p_1q_2q_3 + q_1p_2q_3 + q_1q_2p_3\\
&= 0.329\\
\\
\mathbb{P}(Z=3) &= q_1 q_2 q_3 = 0.612
\end{align*}

```python
z = np.arange(4)
p = np.array([0.003, 0.056, 0.329, 0.612])
z_squared = z**2
expectation = np.dot(z,p)
variance = np.dot(z_squared, p)

print(f"Expected Value M(Z): {expectation}")

print(f"Variance D(Z):       {variance}")
```
