We will assume that the random variable $X \sim Bin(10000, 0.001)$ follows Binomial distribution.


a.

Expected number of errors $\mathbb{E}[X] = np$

b. The probability of observing fewer than four errors, $$\mathbb{P}(X < 4 ) = \sum_{i}^{3} \binom{10000}{i}{} p^iq^{10000-i} $$

c.
$$
\mathbb{P}(X > 2) = 1 - \mathbb{P}(X \leq 2 ) = 1 - \sum_{i}^{2} \binom{10000}{i}{} p^iq^{10000-i}  
$$

```python
# a
p = 0.001
n = 10000
print(f"Expected number of errors is {n*p}")
```

```python
from math import comb

# b
answer_b = 0
q=1-p
for i in range(4):
  answer_b += comb(10000, i)*(p**i)*(q**(n-i))

print(f"The probability of observing fewer than four errors {answer_b:.4f}")

```

```python
# c
answer_c = 1
for i in range(3):
  answer_c -=  comb(10000, i)*p**i*q**(n-i)

print(f"The probability of observing fewer than four errors {answer_c:.4f}")
```
